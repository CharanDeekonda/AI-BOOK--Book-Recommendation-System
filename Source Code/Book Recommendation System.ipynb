```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("books.csv")
df
```

```python
df.columns
```

```python
df.drop(['Unnamed: 0', 'book_id', 'year', 'pages', 
         'average_rating', 'ratings_count', 
         'books_count', 'authors'], axis = 1, inplace = True)
df
```

# Create index_book_dict

```python
index_book_dict = {}
for i in range(len(df)):
    index_book_dict[i] = df.loc[i]['title']
index_book_dict
```

# Create book_index_dict

```python
book_index_dict = {}
for i in range(len(df)):
    book_index_dict[df.loc[i]['title']] = i
book_index_dict
```

# Text Preprocessing

```python
import nltk
nltk.download('stopwords')
nltk.download('wordnet')
import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
wl = WordNetLemmatizer()
```

```python
processed_description = []
for i in range(len(df)):
    description = re.sub(r"http\S+", ' ', df['description'][i])
    description = re.sub('[^a-zA-Z0-9]', ' ', description)
    description = description.lower()
    word_list = description.split()
    word_list = [wl.lemmatize(word) for word in \
                 word_list if not word in stopwords.words('english')]    
    description = ' '.join(word_list)
    processed_description.append(description)
    print(i, end = ' ')
```

```python
df['Processed Description'] = processed_description
df
```

```python
processed_genres = []
for i in range(len(df)):
    # Remove Hyperlinks
    genres = re.sub(r"http\S+", ' ', df['genres'][i])
    # Remove Punctuation Marks and Special Symbols
    genres = re.sub('[^a-zA-Z0-9]', ' ', genres)
    # Lowercase
    genres = genres.lower()
    # Tokenization
    word_list = genres.split()
    # Remove Stopwords and perform Stemming / Lemmatization
    word_list = [wl.lemmatize(word) for word in \
                 word_list if not word in stopwords.words('english')]    
    genres = ' '.join(word_list)
    processed_genres.append(genres)
    print(i, end = ' ')
```

```python
df['Processed Genres'] = processed_genres
df
```

# Text to Vector Conversion

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
processed_description_vectors = cv.fit_transform(processed_description)
df['Processed Description Vectors'] = list(processed_description_vectors.toarray())
df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
processed_genres_vectors = cv.fit_transform(processed_genres)
df['Processed Genres Vectors'] = list(processed_genres_vectors.toarray())
df
```

# Recommendations based on recently read book

```python
book = input('Enter the recently read book: ')
number_of_recommendations = int(input('How many recommendations do you want? '))
```

# Version 1: Recommendations based on similar description

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity_matrix = cosine_similarity(processed_description_vectors)
idx = book_index_dict[book]
similarity_list = similarity_matrix[idx]
lst = []
for i in range(len(similarity_list)):
    lst.append((similarity_list[i], i))
lst.sort(reverse = True)
recommendations = []
for i in range(len(lst)):
    recommendations.append(index_book_dict[lst[i][1]])
recommendations.remove(book)
recommendations[:number_of_recommendations]
```

# Version 2: Recommendations based on similar genres

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity_matrix = cosine_similarity(processed_genres_vectors.toarray())
idx = book_index_dict[book]
similarity_list = similarity_matrix[idx]
lst = []
for i in range(len(similarity_list)):
    lst.append((similarity_list[i], i))
lst.sort(reverse = True)
recommendations = []
for i in range(len(lst)):
    recommendations.append(index_book_dict[lst[i][1]])
recommendations.remove(book)
recommendations[:number_of_recommendations]
```
