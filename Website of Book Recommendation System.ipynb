```python
import numpy as np
import pandas as pd
import pickle
from pywebio.input import *
from pywebio.output import *
from pywebio import start_server
from pywebio.platform.flask import webio_view
from flask import Flask
app = Flask(__name__)

def getStarted():
    # def open_homepage():
    #     run_js('window.open("http://localhost/books", "_blank);')
    
    with open('books_dataframe', 'rb') as f:
        df = pickle.load(f)
    with open('description_similarity_matrix', 'rb') as f:
        description_similarity_matrix = pickle.load(f)
    with open('genres_similarity_matrix', 'rb') as f:
        genres_similarity_matrix = pickle.load(f)
    
    # Create index_book_dict
    index_book_dict = {}
    for i in range(len(df)):
        index_book_dict[i] = df.loc[i]['title']
    
    # Create book_index_dict
    book_index_dict = {}
    for i in range(len(df)):
        book_index_dict[df.loc[i]['title']] = i
    
    # Create book_image_dict
    book_image_dict = {}
    for i in range(len(df)):
        book_image_dict[df.loc[i]['title']] = df.loc[i]['image_url']
    put_html('<html lang="en" dir="ltr"><head><meta charset="utf-8" /><title> Rishabh Bafna </title><style> body {  background-color: white;} </style></head></html>')
    put_markdown('# AI BOOK')
    img = open('aibook.avif', 'rb').read()
    put_image(img, width='100%')
    options_list = list(df['title'])
    book = select("Which is the recently read book by you? ", options = options_list, required = True)
    number_of_recommendations = input("Enter the number of recommendations", type = NUMBER)
    recommendation_system = select('Which type of recommendation system would you prefer?', ['Recommendations on the basis of similar book description', 'Recommendations on the basis of similar genres'])
    put_html('<hr>')
    put_markdown('### Please wait! Your request is being processed. ')
    import time
    put_processbar('bar');
    for i in range(1, 11):
        set_processbar('bar', i / 10)
        time.sleep(0.1)
    if(recommendation_system == 'Recommendations on the basis of similar book description'):
        idx = book_index_dict[book]
        similarity_list = description_similarity_matrix[idx]
        lst = []
        for i in range(len(similarity_list)):
            lst.append((similarity_list[i], i))
        lst.sort(reverse = True)
        recommendations = []
        for i in range(len(lst)):
            recommendations.append(index_book_dict[lst[i][1]])
        recommendations.remove(book)
        final_recommendations_list = recommendations[:number_of_recommendations]
        put_markdown('### Recommendations similar to `%s` book are as follows:' % book)
        for book in final_recommendations_list:
            put_html('<hr>')
            put_markdown("# *`%s`*" % book)
            put_image(book_image_dict[book], width='30%')
    if(recommendation_system == 'Recommendations on the basis of similar genres'):
        idx = book_index_dict[book]
        similarity_list = genres_similarity_matrix[idx]
        lst = []
        for i in range(len(similarity_list)):
            lst.append((similarity_list[i], i))
        lst.sort(reverse = True)
        recommendations = []
        for i in range(len(lst)):
            recommendations.append(index_book_dict[lst[i][1]])
        recommendations.remove(book)
        final_recommendations_list = recommendations[:number_of_recommendations]
        put_markdown('### Recommendations similar to `%s` book are as follows:' % book)
        for book in final_recommendations_list:
            put_html('<hr>')
            put_markdown("# *`%s`*" % book)
            put_image(book_image_dict[book], width='30%')
    # put_button("Back", onclick=open_homepage)
    put_html('<hr>')
    put_html('<h1 style="font-size:20px; text-align:center; color:blue; background-color:white;"> Thank you! :-) <br> </h1>') 
    put_html('<hr>')
app.add_url_rule('/books', 'webio_view', webio_view(getStarted), methods=['GET', 'POST', 'OPTIONS'])
app.run(host='localhost', port=80)

# Check out http://localhost/books
```
